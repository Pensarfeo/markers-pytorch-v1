````python
def get_latest_opset():
    """Return second-most (for maturity) recently supported ONNX opset by this version of torch."""
    return max(int(k[14:]) for k in vars(torch.onnx) if 'symbolic_opset' in k) - 1  # opset
````

````python
dirName = './outputs/Yolov8-cocoTest-noise-brighness-colortwicking-counterExamples-d0-test-crl-asym/epochs-4'
fileName = os.path.join(dirName, 'model.onnx')
try:
    os.mkdir(dirName)
except:
    1
````

````python
inputTensor = x = torch.randn(120, 3, 256, 256, requires_grad=True)
torch_out = model(inputTensor)


# os.mkdir(dirName)
torch.onnx.export(
model,
inputTensor,
fileName,
verbose=False,
opset_version=get_latest_opset(),
do_constant_folding=True,  # WARNING: DNN inference with torch>=1.12 may require do_constant_folding=False
input_names=['images'],
output_names=['class', 'bbox', 'wh', 'xy'],
dynamic_axes={
    "images": {0: 'batch_size'},
    "class": {0: 'batch_size'},
    "bbox": {0: 'batch_size'},
    "xy": {0: 'batch_size'},
    "wh": {0: 'batch_size'}
}
)
````

````python
from onnx2tf import convert
tfSaveFolder = dirName,
convert(
    input_onnx_file_path = fileName,
    output_folder_path = tfSaveFolder[0],
    output_signaturedefs = True
)
````

````python
from tensorflowjs.converters.converter import convert
````

````python
class Args () :
    def __init__(self):
        self.input_format = 'tf_saved_model' 
        self.output_format = 'tfjs_graph_model' 
        self.signature_name = 'serving_default' 
        self.saved_model_tags = 'serve'
        self.input_path = dirName
        self.output_path = os.path.join(dirName, 'js')
        self.show_version = True
args = Args()
convert(args)
````

````python
# This is the official way of doing it! The function call one happened because you hacked this thing

from IPython.display import Markdown, display

tf_model_dir=dirName
tfjs_model_dir=os.path.join(dirName, 'tfjs')

tfjs_convert_command = f"""tensorflowjs_converter
                 --input_format=tf_saved_model 
                 --output_format=tfjs_graph_model 
                 --signature_name=serving_default 
                 --saved_model_tags=serve 
                 "{tf_model_dir}" 
                 "{tfjs_model_dir}"
                 "--quantize_uint8"
                 """
tfjs_convert_command = " ".join(tfjs_convert_command.split())
display(Markdown(f"```bash\n{tfjs_convert_command}\n```"))

print("Exporting TensorFlow SavedModel to TensorFlow.js Graph model...")
conversion_result = %sx $tfjs_convert_command
print("\n".join(conversion_result))
````
