```python

def locTensor(bz, size):
    locTensor = tf.range(0, size, dtype=tf.float32)
    locTensorX = tf.expand_dims(locTensor, 0)
    locTensorX = tf.expand_dims(locTensorX, 2)
    locTensorX = tf.tile(locTensorX, [bz, 1, size])
    locTensorY = tf.transpose(locTensorX, perm=[0, 2, 1])
    locTensor = tf.stack([locTensorX, locTensorY], axis=3)
```
