```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
import os

def plotRun(
    name,
    ylim = 0,
    ymin = 0,
    ma=1,
    iters=1276,
    xi=0,
    xf=0,
    c = [1, 1, 1, 1],
    showLr = True,
    momentum=False,
    fsx = 12,
    fsy = 6
    ):
    # data1 = pd.read_csv('./Yolov8-1Epochs.csv')

    csvPath = os.path.join('.', 'outputs', name, 'output.csv')
    data = pd.read_csv(csvPath)
    colors=[['red','orange', 'deeppink', 'purple', 'deeppink'], ['purple', 'pink']]
    nyp = 1

    withMomentum = momentum and 'momentum' in data.columns
    if withMomentum:
        nyp += 1 
    if showLr:
        nyp += 1
    
    if xf == 0:
        xf = len(data.di)

    di = data.di[xi:xf].rolling(ma).mean()
    ce = data.ce[xi:xf].rolling(ma).mean()
    bboxL1 = data.bboxL1[xi:xf].rolling(ma).mean()
    lossDIOU = data.lossDIOU[xi:xf].rolling(ma).mean()
    

    figsize = (fsx, fsy+3*(nyp-1))
    heightRatios = [2, *([1]*(nyp-1))]
    fig, ax = plt.subplots(nyp, 1, figsize = figsize, height_ratios= heightRatios)

    if not type(ax).__module__ == 'numpy':
        ax = [ax]

    ax[0].plot(list(range(len(di))), di*c[0], color=colors[0][0], label='dice')
    ax[0].plot(list(range(len(di))), ce*c[1], color=colors[0][1], label='ce')
    ax[0].plot(list(range(len(di))), lossDIOU*c[2], color=colors[0][2], label='diou')
    ax[0].plot(list(range(len(di))), bboxL1*c[3], color=colors[0][3], label='bboxL2') 

    ax[0].grid(True)

    ax[0].legend()
    iters = 1835  
    for i in range(len(data.ce[xi:xf])//iters):
        ax[0].axvline((i+1)*iters, 0, 1)
    if ylim:
        ax[0].set_ylim([ymin, ylim])
#     ax1.set_aspect(10)
    #     time.sleep(5)
    # clear_output(wait=False)

    p = 0
    if (showLr):
        p += 1
        ax[p].plot(list(range(len(data.lr[xi:xf]))), data.lr[xi:xf], color=colors[0][0], label='dice', linestyle='dashed')
        ax[p].set_xlim(left=0)
        ax[p].grid(True)
    
    if (withMomentum):
        p += 1
        ax[p].plot(list(range(len(data.momentum[xi:xf]))), data.momentum[xi:xf], color=colors[0][0], label='dice', linestyle='dashed')
        ax[p].set_xlim(left=0)
        ax[p].grid(True)


    plt.show()
    return xf

name = "Yolov8-cocoTest-noise-brighness-colortwicking-counterExamples-d0-test-lr"
iters = 1835
ylim = 0.15
ymin = 0.05
ma = 10
fsy = 3
showLr = True
momentum = False

def xi(n, x = 0):
    return 2710 + round(iters*(n + (x/iters)))

def xf(n, l):
    return 2710 + round(iters*(n + (l/iters)))

plotRun(
    name,
    iters = iters,
    ylim = ylim,
    ma = ma,
    xi = xi(0),
    xf = xf(0, xfp),
    showLr = showLr,
    momentum = momentum,
    c = [1, 1, 1, 1],
    fsy = fsy,
)

# name = "Yolov8-cocoTest-noise-brighness-colortwicking-counterExamples-d0-test-lr"
# ylim = 10

xp = plotRun(
    name,
    iters = iters,
    ylim = ylim,
    ma = ma,
    xi = xi(2),
    xf = xf(2, xfp),

    showLr = showLr,
    momentum = momentum,
    c = [1, 1, 1, 1],
    fsy = fsy,
)

xp = plotRun(
    name,
    iters = iters,
    ylim = ylim,
    ma = ma,
    xi = xi(4),
    xf = xf(4, xfp),

    showLr = showLr,
    momentum = momentum,
    c = [1, 1, 1, 1],
    fsy = fsy,
)

xp = plotRun(
    name,
    iters = iters,
    ylim = ylim,
    ymin = ymin,
    ma = ma,
    xi = xi(6, iters),

    showLr = showLr,
    momentum = momentum,
    c = [1, 1, 1, 1],
    fsy = fsy,
)


xfp = (xp - 2710 - iters*6)
```

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

def plot_results(file='path/to/results.csv', xi=0, xf=None , dir='', segment=False):
    # Plot training results.csv. Usage: from utils.plots import *; plot_results('path/to/results.csv')
    save_dir = Path(file).parent if file else Path(dir)
    if segment:
        fig, ax = plt.subplots(2, 8, figsize=(18, 6), tight_layout=True)
        index = [1, 2, 3, 4, 5, 6, 9, 10, 13, 14, 15, 16, 7, 8, 11, 12]
    else:
        fig, ax = plt.subplots(2, 5, figsize=(12, 6), tight_layout=True)
        index = [1, 2, 3, 4, 5, 8, 9, 10, 6, 7]
    ax = ax.ravel()
    files = list(save_dir.glob('results*.csv'))
    assert len(files), f'No results.csv files found in {save_dir.resolve()}, nothing to plot.'
    for f in files:
        try:
            data = pd.read_csv(f)
            s = [x.strip() for x in data.columns]
            xf = xf if isinstance(xf, int) else len(data.values)
            x = data.values[xi:xf, 0]
            for i, j in enumerate(index):
                y = data.values[xi:xf, j].astype('float')
                # y[y == 0] = np.nan  # don't show zero values
                ax[i].plot(x, y, marker='.', label=f.stem, linewidth=2, markersize=8)
                ax[i].set_title(s[j], fontsize=12)
                # if j in [8, 9, 10]:  # share train and val loss y axes
                #     ax[i].get_shared_y_axes().join(ax[i], ax[i - 5])
        except Exception as e:
            print(f'Warning: Plotting error for {f}: {e}')
    ax[1].legend()
    plt.show()
    fig.savefig(save_dir / 'results.png', dpi=200)
    plt.close()
```
